```python
from types import SimpleNamespace

system_cfg = {'punct':False}

train_cfg = {'lr':1e-5, 'scheduling':False, 'epochs':3, 'print_len':100, 'bsz':4}

system_cfg = SimpleNamespace(**system_cfg)
train_cfg = SimpleNamespace(**train_cfg)

E = ExperimentHandler(system_cfg)
E.train(train_cfg)
```

```python
import numpy as np
import torch

confusion_matrix = np.zeros([len(E.D.act_dict), len(E.D.act_dict)])
for k, batch in enumerate(E.B.batches(E.D.dev), start=1):
    y = E.model(batch.ids, batch.mask)
    loss = E.cross_loss(y, batch.labels)

    y_pred = torch.argmax(y, -1)

    for pred, lab in zip(y_pred, batch.labels):
        confusion_matrix[lab, pred] += 1

    TP = np.diagonal(confusion_matrix)
    label_count = np.sum(confusion_matrix, axis=0)
    pred_count = np.sum(confusion_matrix, axis=1)

    if k>500:
        break
        
precision = TP/pred_count
recall = TP/label_count
F1 = 2*(precision*recall)/(precision+recall)
macro_F1 = np.mean(F1[~np.isnan(F1)])
micro_F1 = np.nansum(label_count*F1/np.sum(label_count))
overall_P = sum(TP)/sum(pred_count)
overall_R = sum(TP)/sum(label_count)

print(f'macro F1: {macro_F1:.3f}   micro F1: {micro_F1:.3f}' \
f'precision: {overall_P:.3f} recall: {overall_R:.3f}')

print(overall_P, overall_R)
print(macro_F1, micro_F1)
```
