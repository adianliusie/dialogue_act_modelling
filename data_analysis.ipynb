```python
from src.helpers.conv_handler import ConvHandler
```

```python
C = ConvHandler(system='bert',punct=True, action=True)
C_filt = ConvHandler(system='bert',punct=False, action=False)
```

```python
conv_stats = {'words':[], 'utts':[], 'turns':[], 'ids':[]}
for conv in C.train:
    conv_stats['ids'].append(len([i for utt in conv.utts for i in utt.ids]))
    conv_stats['words'].append(len([i for utt in conv.utts for i in utt.text.split()]))
    conv_stats['utts'].append(len(conv.utts))
    conv_stats['turns'].append(len(conv.turns))

filt_conv_stats = {'words':[], 'utts':[], 'turns':[], 'ids':[]}
for conv in C_filt.train:
    filt_conv_stats['ids'].append(len([i for utt in conv.utts for i in utt.ids]))
    filt_conv_stats['words'].append(len([i for utt in conv.utts for i in utt.text.split()]))
    filt_conv_stats['utts'].append(len(conv.utts))
    filt_conv_stats['turns'].append(len(conv.turns))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(conv_stats['ids'])
plt.show()

sns.histplot(conv_stats['words'])
plt.show()

sns.histplot(conv_stats['utts'])
plt.show()

sns.histplot(conv_stats['turns'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(filt_conv_stats['ids'])
plt.show()

sns.histplot(filt_conv_stats['words'])
plt.show()

sns.histplot(filt_conv_stats['utts'])
plt.show()

sns.histplot(filt_conv_stats['turns'])
plt.show()
```
